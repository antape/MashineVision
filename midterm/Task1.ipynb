```python
from scipy import ndimage
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
from skimage import color
from skimage import io
from skimage import util

import matplotlib.image as pltimg
img =  io.imread('photo.jpg', as_grey=True)
img = util.random_noise(img, mode='s&p', seed=None, clip=True)




#Filtr A
k = np.array([[1.*1/6,2.*1/6,1.*1/6],[2.*1/6,4.*1/6,2.*1/6],[1.*1/6,2.*1/6,1.*1/6]])
#k = k ** 1/16
flt_1 = ndimage.convolve(img, k, mode='nearest', cval=0.0)


#Filtr B
k = np.array([[0.*1/5,1.*1/5,0.*1/5],[1.*1/5,1.*1/5,1.*1/5],[0.*1/5,1.*1/5,0.*1/5]])
#k = k ** 1/5
flt_2 = ndimage.convolve(img, k, mode='nearest', cval=0.0)


#Filtr G
k = np.array([[-1.,-2.,-1.],[-2.,12.,-2.],[-1.,-2.,-1.]])
flt_4 = ndimage.convolve(img, k, mode='nearest', cval=0.0)




#Filtr D
k = np.array([[-1.,-2.,-1.],[0.,0.,0.],[1.,2.,1.]])
flt_5 = ndimage.convolve(img, k, mode='nearest', cval=0.0)

#Filtr F
flt_6 = ndimage.median_filter(img, 3)




fig = plt.figure(figsize=(35,35))
plt.subplot(711)
plt.title('Original')
plt.imshow(img,cmap='gray')
plt.subplot(712)
plt.title('Filtr A')
plt.imshow(flt_1,cmap='gray')
plt.subplot(713)
plt.title('Filtr B')
plt.imshow(flt_2,cmap='gray')
plt.subplot(714)
plt.title('Filtr G')
plt.imshow(flt_4,cmap='gray')
plt.subplot(715)
plt.title('Filtr D')
plt.imshow(flt_5,cmap='gray')
plt.subplot(716)
plt.title('Filtr F')
plt.imshow(flt_6,cmap='gray')



plt.show()
```
